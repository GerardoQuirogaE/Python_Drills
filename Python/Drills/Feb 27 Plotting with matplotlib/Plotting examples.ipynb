```python
import numpy as np
import matplotlib.pyplot as plt

#create y data with the same length as xs
xs = np.linspace(0, 10, 50)
ys = np.empty(len(xs))
for i in range(len(ys)):
    ys[i] = xs[i]**2 #function here can be whatever we want

plt.plot(xs, ys)
plt.show()
```

```python
plt.scatter(xs, ys)
plt.show()
```

```python
#they don't have to both have the same xs
xs2 = np.linspace(0, 5, 50)
ys2 = np.empty(len(xs2))
for i in range(len(xs2)):
    ys2[i] = 3*np.cos(xs[i])
    
plt.plot(xs, ys, color='k', linestyle=':', label='squared')
plt.plot(xs2, ys2, color='red', linestyle='--', label='linear')
plt.title('Squared')
plt.xlabel('xs')
plt.ylabel('ys')
plt.legend()
plt.show()
```

```python
#slicing arrays
xs2_sliced = xs2[20:30]
ys2_sliced = ys2[20:30] 

plt.plot(xs2, ys2, label='original')
plt.plot(xs2_sliced, ys2_sliced, label='sliced', linestyle='--', linewidth=5)
plt.legend()
plt.show()
```

```python
xs_log = np.linspace(1, 200, 1000)
xs2_log = np.linspace(1, 100, 1000)
ys_log = np.empty(len(xs_log))
ys2_log = np.empty(len(xs_log))

for i in range(len(xs_log)): # create a function that gets large quickly
    ys_log[i] = np.exp(np.sqrt(xs_log[i]))
for i in range(len(xs2_log)): # create a function that gets large quickly
    ys2_log[i] = np.exp(2*xs2_log[i])
    


plt.plot(xs_log, ys_log)
plt.plot(xs2_log, ys2_log, linestyle='--')
plt.show()
```

```python
#That doesn't look useful. We try a log plot instead:

plt.plot(xs_log, ys_log)
plt.plot(xs2_log, ys2_log)
plt.yscale('log')
plt.show()
```
