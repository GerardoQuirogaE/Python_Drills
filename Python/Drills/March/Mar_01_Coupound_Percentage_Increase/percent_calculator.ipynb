```python
import numpy as np
import matplotlib.pyplot as plt
import mplcursors
import matplotlib.ticker as ticker
```

```python
def percent_better_numpy(n=7):
    initial = 100
    increase = 0.01
    
    days = np.arange(1, n + 1)
    values = initial * (1 + increase) ** days
    percent_increase = values - initial
    
    for day, perc in zip(days, percent_increase):
        print(f"% increase for day {day} is {perc:,.3f}%")
    
    print("\nStats:")
    print("--------")
    print(f"For {n} days, compounding {increase*100}% each day:")
    print(f"Percentage increase: {percent_increase[-1]:,.3f}%")
    print(f"Percentage gain over simple interest: "
          f"{percent_increase[-1] - (increase*n*100):,.3f}%")
    

percent_better_numpy(int(input("How many days: ")))
```

```python
def plot_percent_better(n=30):
    initial = 100

    increase1 = 0.01
    increase2 = 0.02
    increase3 = 0.03
    
    days = np.arange(1, n + 1)

    values1 = initial * (1 + increase1) ** days
    percent_increase1 = values1 - initial

    values2 = initial * (1 + increase2) ** days
    percent_increase2 = values2 - initial

    values3 = initial * (1 + increase3) ** days
    percent_increase3 = values3 - initial
    
    plt.figure()
    plt.scatter(days, percent_increase1, label = "1% Better every day")
    plt.scatter(days, percent_increase2, label = "2% Better every day")
    plt.scatter(days, percent_increase3, label = "3% Better every day")
    plt.xlabel("Days")
    plt.ylabel("Percent Increase")
    plt.title("1% Better for 75 Days")
    plt.legend()
    plt.show()

plot_percent_better(75)
```

```python
def plot_percent_better(n=30):

    increase1 = 0.01
    increase2 = 0.02
    increase3 = 0.03
    
    days = np.arange(1, n + 1)

    # Compound percent increase
    compound1 = (1 + increase1) ** days - 1
    compound2 = (1 + increase2) ** days - 1
    compound3 = (1 + increase3) ** days - 1

    percent_increase1 = compound1 * 100
    percent_increase2 = compound2 * 100
    percent_increase3 = compound3 * 100

    simple1 = increase1 * days * 100
    simple2 = increase2 * days * 100
    simple3 = increase3 * days * 100

    gain1 = percent_increase1 - simple1
    gain2 = percent_increase2 - simple2
    gain3 = percent_increase3 - simple3

    # Create figure FIRST
    fig, ax = plt.subplots(figsize=(10,6))

    s1 = ax.scatter(days, percent_increase1, label="1% Better every day")
    s2 = ax.scatter(days, percent_increase2, label="2% Better every day")
    s3 = ax.scatter(days, percent_increase3, label="3% Better every day")

    # Disable scientific notation
    ax.ticklabel_format(style='plain', axis='y')

    # Add commas
    ax.yaxis.set_major_formatter(
        ticker.FuncFormatter(lambda x, pos: f"{x:,.0f}")
    )

    ax.set_xlabel("Days")
    ax.set_ylabel("Percent Increase")

    ax.set_title(
        f"Compounding Growth of 1%, 2%, and 3% Interest\n"
        f"Over {n} Days",
        fontweight="bold"
    )

    ax.legend()

    cursor = mplcursors.cursor([s1, s2, s3], hover=True)

    @cursor.connect("add")
    def on_add(sel):
        index = sel.index
        x = days[index]

        if sel.artist == s1:
            y = percent_increase1[index]
            g = gain1[index]
            p = "1%"
        elif sel.artist == s2:
            y = percent_increase2[index]
            g = gain2[index]
            p = "2%"
        else:
            y = percent_increase3[index]
            g = gain3[index]
            p = "3%"

        # Clean aligned table
        sel.annotation.set_text(
            "\n".join([
                rf"$\bf{{Day:}}$ {x}",
                rf"$\bf{{Improvement\ Rate:}}$ {p}%",
                rf"$\bf{{Compound\ Increase:}}$ {y:,.2f}%",
                rf"$\bf{{Extra\ Gain:}}$ {g:,.2f}%",
                rf"$\bf{{Avg\ Per\ Day:}}$ {g/x:,.2f}%"
            ])
        )

    plt.tight_layout()
    plt.show()

plot_percent_better(int(input("How many days: ")))
```

```python
def plot_percent_better(n=30):

    increase1 = 0.01
    increase2 = 0.02
    increase3 = 0.03
    
    days = np.arange(1, n + 1)

    # Compound percent increase
    compound1 = (1 + increase1) ** days - 1
    compound2 = (1 + increase2) ** days - 1
    compound3 = (1 + increase3) ** days - 1

    percent_increase1 = compound1 * 100
    percent_increase2 = compound2 * 100
    percent_increase3 = compound3 * 100

    simple1 = increase1 * days * 100
    simple2 = increase2 * days * 100
    simple3 = increase3 * days * 100

    gain1 = percent_increase1 - simple1
    gain2 = percent_increase2 - simple2
    gain3 = percent_increase3 - simple3

    fig, ax = plt.subplots(figsize=(10,6))

    s1 = ax.scatter(days, percent_increase1, label="1% Better every day")
    s2 = ax.scatter(days, percent_increase2, label="2% Better every day")
    s3 = ax.scatter(days, percent_increase3, label="3% Better every day")

    ax.ticklabel_format(style='plain', axis='y')

    ax.yaxis.set_major_formatter(
        ticker.FuncFormatter(lambda x, pos: f"{x:,.0f}")
    )

    ax.set_xlabel("Days")
    ax.set_ylabel("Percent Increase")

    ax.set_title(
        f"Compounding Growth of 1%, 2%, and 3% Interest\n"
        f"Over {n} Days",
        fontweight="bold"
    )

    ax.legend()

    cursor = mplcursors.cursor([s1, s2, s3], hover=True)

    @cursor.connect("add")
    def on_add(sel):
        index = sel.index
        x = days[index]

        if sel.artist == s1:
            y = percent_increase1[index]
            g = gain1[index]
            p = "1%"
        elif sel.artist == s2:
            y = percent_increase2[index]
            g = gain2[index]
            p = "2%"
        else:
            y = percent_increase3[index]
            g = gain3[index]
            p = "3%"

        # Clean aligned table
        label_width = 22

        rows = [
            ("Day", f"{x}"),
            ("Improvement Rate", p),
            ("Compound Increase", f"{y:,.2f}%"),
            ("Extra Gain", f"{g:,.2f}%"),
            ("Avg Per Day", f"{g/x:,.2f}%"),
        ]

        table_text = "\n".join(
            f"{label:>{label_width}} : {value:<12}"
            for label, value in rows
        )

        sel.annotation.set_text(table_text)
        sel.annotation.set_fontfamily("monospace")

    plt.tight_layout()
    plt.show()


plot_percent_better(int(input("How many days: ")))
```
